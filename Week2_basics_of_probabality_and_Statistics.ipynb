In a cross between two heterozygous individuals (AaÃ—Aa), the sample space is S={AA,Aa,aA,aa}. The total size of S is 4.

Section 1: Basic Probability & Sample Space<br>

Probability is the language of chance. In biology, this is key for understanding genetics, disease risk, and experimental outcomes.

Exercise 1.1: Rolling the Genetic Dice ðŸŽ²<br>

Let's simulate the outcomes of a simple genetic cross. Imagine crossing two heterozygous plants (RrÃ—Rr), where 'R' (Red) is dominant and 'r' (white) is recessive.

The possible outcomes (Sample Space) are {RR,Rr,rR,rr}.

The total number of outcomes is 4.

Goal: Calculate the probability of a white phenotype

```python
# 1. Define the sample space size
Total_Outcomes = 4

# 2. Define the number of outcomes for the desired event (white phenotype, rr)
Event_White = 1

# 3. Calculate the probability
Probability_White = Event_White / Total_Outcomes

# 4. Print the result
print(f"The probability of a white phenotype (rr) is: {Probability_White}")
```

Section 2: Binomial Distribution (Discrete Counts)<br>

The Binomial Distribution models the probability of getting a certain number of successes (k) in a fixed number of independent trials (n), where there are only two outcomes (e.g., infected/not infected, mutated/normal).

Key parameters:

n: Number of trials (e.g., 20 mice in a study).

p: Probability of success (e.g., 0.3 chance of survival).

Exercise 2.1: Simulating Drug Efficacy ðŸ’Š
A new drug has a 30% chance of curing a specific parasite in a single patient. You are testing the drug on 20 patients. We want to simulate how often you might see 6 successful cures.

n=20 (trials/patients)

p=0.3 (probability of cure/success)

k=6 (desired successes)

```python
import numpy as np
from scipy.stats import binom

# 1. Define the Binomial parameters
n = 20    # Number of patients
p = 0.3   # Probability of cure

# 2. Calculate the probability of EXACTLY 6 cures (k=6)
k = 6
P_k6 = binom.pmf(k, n, p)


print(f"Probability of EXACTLY 6 cures: {P_k6:.4f}")
```

Exercise 2.1: Analyzing Leaf Length Data ðŸŒ¿<br>

A biologist measures the length (in cm) of 15 leaves from a new plant species.

```python
import numpy as np
from scipy import stats

# Leaf length data (in cm)
leaf_lengths = np.array([4.2, 5.5, 4.8, 6.1, 5.5, 5.0, 4.9, 5.7, 5.5, 6.2, 4.5, 5.3, 5.1, 4.7, 5.6])

# Calculate Mean (Average)
mean_length = np.mean(leaf_lengths)

# Calculate Median (The middle value)
median_length = np.median(leaf_lengths)

# Calculate Mode (Most frequent value)
# stats.mode returns the mode and its count
mode_result = stats.mode(leaf_lengths, keepdims=True)
mode_length = mode_result.mode[0]

print(f"Mean Leaf Length: {mean_length:.2f} cm")
print(f"Median Leaf Length: {median_length:.2f} cm")
print(f"Mode Leaf Length: {mode_length:.2f} cm")
```

Exercise 2.2: Calculating Variability for Leaf Length

```python
# Data from Exercise 2.1
# leaf_lengths = np.array([4.2, 5.5, 4.8, 6.1, 5.5, 5.0, 4.9, 5.7, 5.5, 6.2, 4.5, 5.3, 5.1, 4.7, 5.6])

# Calculate Range
data_range = np.max(leaf_lengths) - np.min(leaf_lengths)

# Calculate Variance
variance_length = np.var(leaf_lengths, ddof=1) # ddof=1 for sample variance

# Calculate Standard Deviation
std_dev_length = np.std(leaf_lengths, ddof=1) # ddof=1 for sample std dev

print(f"Range of Leaf Lengths: {data_range:.2f} cm")
print(f"Variance (Sample): {variance_length:.2f} cmÂ²")
print(f"Standard Deviation (Sample): {std_dev_length:.2f} cm")
print("\nInterpretation: On average, a leaf's length is about 0.58 cm away from the mean.")
```

Visualize the Bionomial Distribution

```python
import matplotlib.pyplot as plt

# Define possible number of cures (0 to 20)
x = np.arange(0, n + 1)

# Calculate the probability for each possible number of cures
binomial_probs = binom.pmf(x, n, p)

# Create the plot
plt.figure(figsize=(10, 5))
plt.bar(x, binomial_probs, color='skyblue')
plt.title(f'Binomial Distribution for Drug Efficacy (n={n}, p={p})')
plt.xlabel('Number of Successful Cures (k)')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(axis='y', alpha=0.5)

# Highlight the expected mean (n*p)
expected_mean = n * p
plt.axvline(expected_mean, color='red', linestyle='dashed', linewidth=2, label=f'Expected Mean = {expected_mean}')
plt.legend()
plt.show()
```

Section 3: Normal Distribution (Continuous Data)<br>

The Normal Distribution (Bell Curve) models continuous biological measurements like height, weight, and blood pressure. It is defined by two parameters:

Mean (Î¼): The center/average of the data.

Standard Deviation (Ïƒ): The spread of the data.

Exercise 3.1: Human Height Analysis
Assume the height of adult males is normally distributed with:

Mean (Î¼) =175 cm

Standard Deviation (Ïƒ) =7 cm

Goal: Calculate the probability that a randomly chosen male is taller than 185 cm.

```python
from scipy.stats import norm

# 1. Define the Normal Distribution parameters
mean_height = 175
std_dev_height = 7

# 2. Define the threshold (x)
x_threshold = 185

# 3. Calculate the probability of being TALLER than 185 cm.
# norm.cdf(x) gives P(X <= x).
# P(X > x) = 1 - P(X <= x)
P_taller_than_185 = 1 - norm.cdf(x_threshold, mean_height, std_dev_height)

print(f"The probability of being taller than 185 cm is: {P_taller_than_185:.4f} (or {P_taller_than_185*100:.2f}%)")
```

Exercise 3.2: The Empirical Rule
The Empirical Rule states that for a Normal Distribution:

âˆ¼68% of data falls within Â±1Ïƒ of the mean.

âˆ¼95% of data falls within Â±2Ïƒ of the mean.

âˆ¼99.7% of data falls within Â±3Ïƒ of the mean.

Let's calculate the range that contains 95% of adult male heights.

```python
# Use the same parameters: mean=175, std_dev=7

# Calculate the bounds for 95% (mean +/- 2 * std_dev)
Lower_Bound_95 = mean_height - (2 * std_dev_height)
Upper_Bound_95 = mean_height + (2 * std_dev_height)

print(f"Approximately 95% of adult male heights fall between {Lower_Bound_95} cm and {Upper_Bound_95} cm.")
```

```python
# Generate x values (heights)
x_norm = np.linspace(mean_height - 4*std_dev_height, mean_height + 4*std_dev_height, 100)

# Calculate the probability density function (PDF)
pdf_norm = norm.pdf(x_norm, mean_height, std_dev_height)

plt.figure(figsize=(10, 5))
plt.plot(x_norm, pdf_norm, color='darkblue', linewidth=2)
plt.title(f'Normal Distribution of Human Height (Î¼={mean_height}, Ïƒ={std_dev_height})')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')

# Shade the 95% region (161 cm to 189 cm)
fill_x = np.linspace(Lower_Bound_95, Upper_Bound_95, 100)
fill_y = norm.pdf(fill_x, mean_height, std_dev_height)
plt.fill_between(fill_x, fill_y, color='red', alpha=0.3, label='95% Range')

plt.axvline(mean_height, color='black', linestyle='--', label='Mean (175 cm)')
plt.legend()
plt.show()
```
